```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# Scartter plot (diagramas de dispersion)
from cProfile import label
from turtle import color, title


x = np.arange(1, 20)
y= x**2
y1= x**2 + 5*x

plt.scatter(x, y, color="red", marker="s", label="x^2")
plt.scatter(x, y1, color="black", marker="s", label="x^2 +5x")
plt.legend(title="Funciones")
plt.title("Tipo de Funciones", fontsize=20)
plt.xlabel("x", color="green")
plt.ylabel("y")
plt.grid()
plt.text(5, 300, "Este Texto", color="blue", fontsize=12)
#plt.xlim(0,5)

## agregar las coordenadas de los marcadores en texto
for val_x in x:
    plt.text(val_x, val_x**2, f"({val_x}, {val_x})")
```

```python
# Histograma
media = 20
std = 3
numero = np.random.normal(media, std, 1000)
plt.hist(numero, bins=20,)
plt.ylabel("Frecuencia")
```
